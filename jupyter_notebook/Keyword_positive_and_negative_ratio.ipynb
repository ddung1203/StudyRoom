```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import collections
from konlpy.tag import Mecab
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
from PIL import Image
from eunjeon import Mecab
mecab = Mecab()
```

```python
data = pd.read_csv('D:\polo_main.csv')

# 중복 제거
data.drop_duplicates(subset = ['comment']) 
# 정규 표현식 수행
data['comment'] = data['comment'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") 
# 공백은 Null 값으로 변경
data['comment'].replace('', np.nan) 
# Null 값 제거
data = data.dropna(how='any') 

stopwords = ['도', '는', '다', '의', '가', '이', '은', '한', '에', '하', '고', '을', '를', '인', '듯', '과', '와', '네', '들', '듯', '지', '임', '게']
data['label'] = np.select([data.star > 2], [1], default=0)
```

```python
li=[]
i=0
for j in data['comment'].to_list():
    if str(data['label'][i]) == '1':
        a=j+' '+'긍정'
        li.append(a)
    elif str(data['label'][i]) == '0':
        b=j+' '+'부정'
        li.append(b)
    elif str(data['label'][i]) == '-1':
        c=j+' '+'중립'
        li.append(c)
    i=i+1
```

```python
lis=[]
for i in li:
    lis.append(mecab.nouns(i))
liss=sum(lis, [])
```

```python
x=[]
for i in lis:
    for j in i:
        if j == '실망':
            x.append(i)
```

```python
qwe=[]
asd=[]
zxc=[]
for i in x:
    for j in i:
        if j == '긍정':
            qwe.append(i)
        elif j == '부정':
            asd.append(i)
        elif j == '중립':
            zxc.append(i)
```

```python
import collections
positive=collections.Counter(data['label'].to_list())[1]
negative=collections.Counter(data['label'].to_list())[0]
middle=collections.Counter(data['label'].to_list())[-1]
```

```python
print('전체 긍정 비율: '+str(round(positive/(positive+negative+middle)*100, 2)))
```

```python
print('키워드 긍정 비율: '+str(round(len(qwe)/(len(asd)+len(qwe)+len(zxc))*100, 2)))
```
